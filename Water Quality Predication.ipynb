### Random Forset Regression Model

```python
pollutants = ['O2', 'NO3', 'NO2', 'SO4', 'PO4', 'CL']
```

```python
# drop the missing values using dropna()
df = df.dropna(subset=pollutants)
df.head()
```

```python
#  Checking for Null Values 
df.isnull().sum()
```

```python
# Feature and target selection where Feature are independent variable and Target are dependent variable
X = df[['id', 'year']]
y = df[pollutants]
# Here x is  Independent variables and y is dependent variable
```

```python
#  Encoding the Categorical Variable - Converts id  into one-hot encoded columns
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
# Splitting the Data into Train and Test 
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Model Training using MultiOutput Regressor with Random forest Regressor
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, y_train)
```

```python
#  Model Prediction
y_pred = model.predict(X_test)
```

```python
# Model Evaluation
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

```python
# Doing Prediction by taking an Example like taking specific station and year.
station_id = '19'
year_input = 2023

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```

```python
# Saving the Model
import joblib
joblib.dump(model, 'pollution_model.pkl')
joblib.dump(X_encoded.columns.tolist(), "model_columns.pkl")
print('Model and cols structure are saved!')
```

### XGBoost Model

```python
# Doing the Model Training using MultiOutputRegressor with XGBoost
from xgboost import XGBRegressor
```

```python
# Training the model with XGBoost model
model = MultiOutputRegressor(XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42))
model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
print("Model Performance:")
for i, col in enumerate(pollutants):
    print(f"{col}: MSE = {mean_squared_error(y_test.iloc[:, i], y_pred[:, i]):.3f}, RÂ² = {r2_score(y_test.iloc[:, i], y_pred[:, i]):.3f}")
```
